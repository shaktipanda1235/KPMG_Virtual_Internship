```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Cleaned_Data')
df.head()
```

```python

def past_bin(r):
    if r<10:
        return 0
    elif r<20:
        return 1
    elif r<30:
        return 2
    elif r<40:
        return 3
    elif r<50:
        return 4
    elif r<60:
        return 5
    elif r<70:
        return 6
    elif r<80:
        return 7
    elif r<90:
        return 8
    else:
        return 9

df['past_3_years_purchase']=df.past_3_years_bike_related_purchases.apply(past_bin)

df=df.drop('past_3_years_bike_related_purchases',1)


def ten_bin(r):
    if r<5:
        return 0
    elif r<10:
        return 1
    elif r<15:
        return 2
    else:
        return 3

df['Tenure']=df.tenure.apply(ten_bin)

df=df.drop('tenure',1)


X=df.drop('Important_Customer',1)
y=df['Important_Customer']
```

```python
X.head()
```

```python
sns.countplot(df.State,hue=df.Important_Customer)
```

```python
from xverse.transformer import WOE
clf=WOE()
```

```python
clf.fit(X,y)
X_woe=clf.transform(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import StackingClassifier
```

```python
base_learners = [
                 ('KNN_model_1', KNeighborsClassifier(n_neighbors=18,algorithm='ball_tree',p=1.3)),
                 ('KNN_model_2', KNeighborsClassifier(n_neighbors=18))]


stack_model = StackingClassifier(estimators = base_learners, final_estimator = XGBClassifier())
stack_model.fit(X_woe,y)

```
